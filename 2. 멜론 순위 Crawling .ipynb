# 멜론 노래 순위 정보 클롤링①

```python
# 예제 2-51 크롬드라이버 실행
from selenium import webdriver
driver = webdriver.Chrome('/usr/local/bin/chromedriver')
```

```python
# 예제 2-52 멜론 인기차트 웹페이지 접속하기
url = 'http://www.melon.com/chart/index.htm'
driver.get(url)
```

```python
# html다운로드 및 BeautifulSoup으로 읽기

from bs4 import BeautifulSoup
html = driver.page_source
soup = BeautifulSoup(html,'html.parser')
soup
```

```python
songs = soup.select('tr')[1:]  #1부터 곡 정보가 담겨져있음.
print(len(songs))
song = songs[0]
```

```python
# 첫번째곡의 제목 찾기
```

```python
title = song.select('span>a')
print(title[0].text)
len(title)
```

```python
# 예제 2-58 곡 제목 찾기 3
title = song.select('div.ellipsis.rank01 > span > a')[0].text
print(len(title))
print(title)
```

```python
# 가수 찾기
singer = song.select('div.ellipsis.rank02 > a')[0].text
print(len(singer))
print(singer)
```

```python
song_data = []
rank = 1
for song in songs:
    title = song.select('div.ellipsis.rank01 > span> a')[0].text
    singer = song.select('div.ellipsis.rank02 > span > a')[0].text
    print(title,singer,sep = '|')
    song_data.append(['Melon',rank,title,singer])
    rank = rank+1
    
```

```python
import pandas as pd
columns = ['서비스','순위','타이틀','가수']
pd_data = pd.DataFrame(song_data,columns = columns)
pd_data.head()
```

```python
# 엑셀로 저장하기
pd_data.to_excel('./melon.xlsx', index = False, engine = 'openpyxl')
```
